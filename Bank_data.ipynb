```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter

import plotly.express as px
import streamlit as st
```

```python
#clean moon_modus file first
data=pd.read_csv('assignment_2_fraud_moon_modus.tsv',sep='\t')
data.info()
data.isnull().sum()
```

```python
data['date'] = pd.to_datetime(data['date'])
data['Month']=pd.DatetimeIndex(data['date']).month
data['Year']=pd.DatetimeIndex(data['date']).year
data['Quarter']=pd.DatetimeIndex(data['date']).quarter
data['Date'] = pd.to_datetime(data['date'], format='%d-%m').dt.strftime('%Y-%m')

data
```

```python
data=data.sort_values("date",ascending=True).reset_index(drop=True)
```

```python
data['moon'].unique()
```

```python
def conditions(s):
    if (s['moon'] =='GA'):
        return 'Ganymedes'
    elif  (s['moon'] == 'CA'):
        return 'Callisto'
    elif  (s['moon'] =='EU'):
           return 'Europa'
    else:
           return 'IO'
        
data['Moon_name'] = data.apply(conditions, axis=1)
data
```

```python
#clean Turnover file
turnover=pd.read_csv('assignment_2_moon_turnover.csv')
turnover.info()
turnover.isnull().sum()
```

```python
copy=turnover.melt(id_vars=["moon"],var_name='Time',value_name='Turnover')
copy
```

```python
copy.info()
copy.isnull().sum()
```

```python
copy['Time'] = pd.to_datetime(copy['Time'])
copy['Quarter'] = copy['Time'].dt.quarter
copy['Year'] = copy['Time'].dt.year
copy['Month'] = copy['Time'].dt.month
copy['Date'] = pd.to_datetime(copy['Time'], format='%d-%m').dt.strftime('%Y-%m')

copy
```

```python
df3 = copy.groupby(['Year','Month','moon','Date'],as_index=False).agg({'Turnover':'sum'}).reset_index(drop=True)
df3
```

```python
df3.info()
```

```python
df4=data.groupby(['Year','Month','Moon_name','Date'],as_index=False).agg({'fraud':'sum'}).reset_index(drop=True)
df4
```

```python
df5 = pd.merge(df4, df3[["Year","Month","moon","Turnover"]] ,
              how="inner", 
              left_on=["Year","Month","Moon_name"], right_on=["Year","Month","moon"]
            ).drop(columns = ['moon']).reset_index(level=0, drop=True)
df5
```

```python
df5['fraud_share'] = (df5['fraud'] / df5['Turnover'])*10000
df5
```

```python
df5=df5.sort_values(["Moon_name","Date"],ascending=True)
df5
```

```python
df5.to_excel("Output.xlsx")
```

# We can also do the moving average and TotalYTD in python as following (This is the 2nd solution - for this exercise I do all calculation in powerbi instead since the data is quite small)

```python
df5copy=df5.copy()
```

```python
#moving average
df5copy['Rolling_Fraud'] = df5copy.reset_index(drop=True).groupby(["Moon_name"])['fraud'].rolling(window=12,min_periods=1).mean().reset_index(level=0, drop=True)
df5copy['Rolling_Turnover'] = df5copy.reset_index(drop=True).groupby(["Moon_name"])['Turnover'].rolling(window=12,min_periods=1).mean().reset_index(level=0, drop=True)
df5copy['Rolling_FraudShare'] = df5copy.reset_index(drop=True).groupby(["Moon_name"])['fraud_share'].rolling(window=12,min_periods=1).mean().reset_index(level=0, drop=True)
df5copy
```

```python
#YTD
df5copy['YTDFraud'] = df5copy.groupby(['Year','Moon_name'])['fraud'].cumsum()
df5copy['YTDTurnover'] = df5copy.groupby(['Year','Moon_name'])['Turnover'].cumsum()
df5copy['YTDFraud_share'] = df5copy.groupby(['Year','Moon_name'])['fraud_share'].cumsum()
df5copy
```
